```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(Independent_Variable.reshape(-1,1),Dependent_Variable)
```

```python
y_pred = reg.predict(Independent_Variable.reshape(-1,1))
```

```python
plt.figure(figsize=(12, 10))
plt.scatter(Independent_Variable, Dependent_Variable,  color='blue')
plt.plot(Independent_Variable.reshape(-1,1), y_pred, color='red')
plt.xlabel('Independent Variable/Input Parameter')
plt.ylabel('Dependent Variable/ Output Parameter')
plt.show()
```
